```python
# #Kaunse loan amount category me borrowers zyada default kar rahe hain?


plt.figure(figsize=(10, 5))

# Categorize loan amount into groups
df['loan_amount_group'] = pd.cut(df['amount'], bins=[0, 5000, 15000, df['amount'].max()],
                                 labels=['Low', 'Medium', 'High'])

# Count plot with custom palette
sns.countplot(x='loan_amount_group', hue='default', data=df, palette=["#008080", "#ff7f50"])

# Title and labels
plt.title("Loan Amount Group vs Default Rate")

# Add grid
plt.grid(axis='y', linestyle='--')

# Show the plot
plt.show()
```

Observation : Lower loan amounts lead to higher default rates.

```python
#Kis age group ke borrowers ka default rate zyada hai?

# Age Group vs Default Rate: Understanding Default Trends Across Age Categories
plt.figure(figsize=(10, 5))

# Categorize age into groups
df['age_group'] = pd.cut(df['age'], bins=[18, 30, 50, df['age'].max()], 
                         labels=['Young', 'Middle', 'Senior'])

# Count plot with custom palette
sns.countplot(x='age_group', hue='default', data=df, palette=["#008080", "#ff7f50"])

# Title and labels
plt.title("Age Group vs Default Rate: Understanding Default Trends Across Age Categories")

# Add grid
plt.grid(axis='y', linestyle='--')

# Show the plot
plt.show()

```

Observation : Young borrowers (18-30) have the highest default rate.

Conclusion: 

From the analysis, we see that loan defaults are influenced by many factors. People with good credit history, own houses, and skilled jobs are also defaulting, which is unexpected. Furniture/Appliances loans, high loan amounts, and multiple existing loans increase the risk of default. Young borrowers (18-30) and those without a phone also default more. These insights can help banks make better lending decisions and reduce risk.

```python

from sklearn.preprocessing import OneHotEncoder

encoder=OneHotEncoder(sparse_output=False)

encoded_data=encoder.fit_transform(df[['purpose','other_credit','housing', 'job', 'phone','default']])

df2=pd.DataFrame(encoded_data, columns=encoder.get_feature_names_out(['purpose','other_credit','housing', 'job','phone','default']))

result=pd.concat([df,df2], axis=1)
result
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()

df['credit_history_LabelEncoded'] = label_encoder.fit_transform(df['credit_history'])

df
```

```python
result=pd.get_dummies(df,dtype=int,drop_first=True)
result
```

```python
X =result.drop('default_yes',axis=1)  # Independent variables
Y = result['default_yes']  # Target variable
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
print('Shape of X_train is -' , x_train.shape)
print('Shape of X_train is -' , x_train.shape)
print('Shape of y_train is -' , y_train.shape)
print ('Shape of y_train is -' , y_train.shape)
```

```python

from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(x_train, y_train)
```

```python
from sklearn.metrics import accuracy_score

y_pred = model.predict(x_test)

# Checking accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy Score:", accuracy)

```
